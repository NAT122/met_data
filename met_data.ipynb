```python
conteo_cultura = df_met['Culture'].value_counts()


# Preparar datos para el gráfico de torta
labels = conteo_cultura.index
sizes = conteo_cultura.values

conteo_cultura.plot(kind='bar', figsize=(10, 6))
plt.title('Relación entre tipo de calle y ocurrencia en cruce')
plt.xlabel('Tipo de Calle')
plt.ylabel('Cantidad de Incidentes')
plt.legend()
plt.show()
```

```python
filtered_cultures = df_met['Culture'].value_counts()
filtered_cultures = filtered_cultures[filtered_cultures > 100]

# Crear un gráfico de torta
plt.figure(figsize=(10, 8))
filtered_cultures.plot.pie(autopct='%1.1f%%', startangle=90, cmap='tab20', legend=False)

# Personalización del gráfico
plt.title("Distribución de Culturas (Frecuencia > 1)")
plt.ylabel("")  # Para ocultar la etiqueta de la columna
plt.show()
```

```python
filtered_cultures = df_met['Culture'].value_counts()
filtered_cultures = filtered_cultures[filtered_cultures > 100]

# Crear un gráfico de barras
plt.figure(figsize=(12, 6))
filtered_cultures.plot(kind='bar', color='skyblue', edgecolor='black')

# Personalización del gráfico
plt.title("Frecuencia de Culturas (Frecuencia > 1)", fontsize=16)
plt.xlabel("Culturas", fontsize=12)
plt.ylabel("Frecuencia", fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```
