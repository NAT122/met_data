```python
# Normalizar la columna Country
def normalize_country(column):
    def clean_entry(entry):
        if pd.isna(entry) or entry == '':
            return ''
        
        # Dividir la entrada por '|', eliminar duplicados manteniendo el orden
        countries = entry.split('|')
        cleaned_countries = []
        for country in countries:
            country = country.strip()  # Eliminar espacios extra
            if country not in cleaned_countries:  # Evitar duplicados
                cleaned_countries.append(country)
        
        # Reunir los países en el formato original
        return '|'.join(cleaned_countries)
    
    return column.apply(clean_entry)

# Aplicar la normalización
```

```python
def correct_typo_simple(country, country_list, threshold=80):
    """
    Corrige errores ortográficos de nombres de países basándose en una lista válida de países.
    
    Args:
    country (str): Nombre del país a corregir.
    country_list (list): Lista de nombres válidos de países.
    threshold (int): Puntaje mínimo de similitud para aceptar una corrección.
    
    Returns:
    str: Nombre del país corregido o el nombre original si no supera el umbral.
    """
    # Buscar la mejor coincidencia utilizando fuzzywuzzy
    match = process.extractOne(country, country_list)
    
    # Validar si el puntaje supera el umbral
    if match and match[1] >= threshold:
        return match[0]  # Devolver el nombre corregido
    else:
        return country  # Devolver el original si no hay coincidencia adecuada
```
