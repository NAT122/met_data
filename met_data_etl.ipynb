Normalizacion y corrección

```python
def correct_typo_simple(country, country_list, threshold=70):
    """
    Corrige errores ortográficos de nombres de países si están en la lista de países válidos.
    
    Args:
        country (str): Nombre del país a corregir.
        country_list (list): Lista de nombres válidos de países.
        threshold (int): Umbral de similitud para aceptar la corrección.

    Returns:
        str: Nombre del país corregido o el nombre original si no se cumple la condición.
    """
    # Ignorar valores "No data" explícitamente
    if country == "No data":
        return country

    # Buscar coincidencias solo si el valor no está ya en la lista
    if country in country_list:
        return country

    # Aplicar corrección si el país no es válido
    match = process.extractOne(country, country_list, score_cutoff=threshold)
    return match[0] if match else country


```

```python
# Aplicar la corrección
df_simples['Normalized_Country'] = df_simples['Normalized_Country'].apply(
    lambda x: correct_typo_simple(x, country_list, threshold=70)
)


```

Acomodar caracteres especiales para transformar en csv 

```python


# Función para buscar caracteres ilegales
def find_illegal_chars(df):
    illegal_pattern = re.compile(r"[\x00-\x1F]")
    for col in df.columns:
        for i, val in enumerate(df[col]):
            if isinstance(val, str) and illegal_pattern.search(val):
                print(f"Carácter ilegal en la columna '{col}', fila {i + 1}: {val}")

find_illegal_chars(df_met)
```

```python
# Reemplazar caracteres ilegales en todo el DataFrame
def clean_text(value):
    if isinstance(value, str):
        # Eliminar caracteres ilegales (ASCII 0-31, excepto \n y \t si son necesarios)
        value = re.sub(r"[\x00-\x08\x0B-\x1F]", "", value)
        # Opcional: Reemplazar saltos de línea por espacios
        value = value.replace("\n", " ").replace("\r", "")
    return value
```

```python

df_met = df_met.applymap(clean_text)
```
